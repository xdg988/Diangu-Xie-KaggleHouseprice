## Handle missing values

### NA values

There are three categories missing values we should consider. 

*  According to the data discription, some features that have missing values can be filled with 'None' or 0. Then we will replace numeric features missing values with zero and categorical features missing values with 'None'.

```python
cat_features1={'BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','BsmtQual','GarageCond','GarageFinish','GarageQual','GarageType','Alley'
              ,'FireplaceQu','Fence','MiscFeature','PoolQC','MasVnrType'}
num_features1={'BsmtFinSF1','BsmtFinSF2','BsmtFullBath','BsmtHalfBath','BsmtUnfSF','TotalBsmtSF','GarageArea','GarageCars',
               'MasVnrArea'}
for c in cat_features1:
    all_data[c]=all_data[c].fillna('None')
    
for n in num_features1:
    all_data[n]=all_data[n].fillna(0)
```

* For other missing values, we cannot get useful information about them. Therefore, we will replace numeric features missing values with median and categorical features missing values with mode.

```python
cat_features2={'Functional','Electrical','Exterior1st','Exterior2nd','KitchenQual','MSZoning','SaleType','Utilities'}
for c in cat_features2:
    all_data[c]=all_data[c].fillna(all_data[c].mode()[0])
#num_features2={'GarageYrBlt'}
all_data['GarageYrBlt']=all_data['GarageYrBlt'].fillna(all_data['GarageYrBlt'].median())
```

* However, we could notice that feature 'LotFrontage' has lots of missing value. Therefore, we group them by feature 'Neighborhood' and then  use median values in each group to fill them.

```python
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

### Features convert

From data description and pictures above, we start to make some transformation between numerical features and categorical features.
What we should be careful is that it is not a good idea to convert too many features between categorical features and numerical features. Because it will easily lose lots of information in raw training and testing data.

 I got this idea from the [link](#https://datascience.stackexchange.com/questions/9202/categorical-and-ordinal-feature-data-representation-in-regression-analysis/9211#9211) and it said:
 
"One hot encoding can learn the relationship between the ordinal values more finely, but throws out the information that the variables are related. Similarly, with insufficient data it is more likely to overfit.
Ordinal variables lessen those problems but at the cost of forcing you to define the interval. There are a number of methods for defining the values of your ordinal variables, like rologit."

* Categorical features to numerical features

```python
new_features2={'MSSubClass','MoSold','YrSold'}
for n in new_features2:
    all_data[n]=all_data[n].astype(str)
```

```python
cat_features2={'FireplaceQu', 'BsmtQual', 'BsmtCond', 'GarageQual', 'GarageCond', 'ExterQual', 'ExterCond','HeatingQC', 'PoolQC', 
               'KitchenQual'}
for c in cat_features2:
    lbl=LabelEncoder()
    lbl.fit(['None','No','Po','Fa','TA','Gd','Ex']) 
    all_data[c] = lbl.transform(list(all_data[c].values))

all_data = all_data.replace({
                "BsmtFinType1": {'None':0 ,'Unf':1 ,'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6},
                "BsmtFinType2": {'None':0 ,'Unf':1 ,'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6},
                "BsmtExposure" : {"None":0,"No" : 1, "Mn" : 2, "Av": 3, "Gd" : 4},
                            })
```

### Skewed features

As the comments in this [kernel](#https://www.kaggle.com/apapiu/regularized-linear-models) says:

"I log transformed certain features for which the skew was > 0.75. This will make the feature more normally distributed and this makes linear regression perform better - since linear regression is sensitive to outliers. Note that if I used a tree-based model I wouldn't need to transform the variables."

We need to make the numerical features more normally distributed.

```python
num_f2=all_data.dtypes[all_data.dtypes!=object].index
skewed=all_data[num_f2].skew()
skewed=skewed[skewed>0.75]
skewed_feat=skewed.index

all_data[skewed.index]=np.log1p(all_data[skewed.index])
```

Of course, Box Cox transformation is also a good idea. In order to be consistent with the way of transforming 'SalePrice', we decide to use log transformation. ### Evaluation

We will use Root-Mean-Sqaured-Error(RMSE) between the logarithm of the predicted value and the logarithm of the observed sales price.

```python
def cv_score(model):
    mse=-cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
    rmse=np.sqrt(mse)
    return rmse
```

### Bagging & Stacking model

I have read some great notebooks about this part. Some ideas and code details can be found in this [kernel](#https://www.kaggle.com/serigne/stacked-regressions-top-4-on-leaderboard).

In this section, we will use model ensemble to improve our base models.

* Averaging base models

```python
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    def __init__(self, models):
        self.models = models
        
    # we define clones of the original models to fit the data in
    def fit(self, X, y):
        self.models_ = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models_:
            model.fit(X, y)

        return self
    
    #Now we do the predictions for cloned models and average them
    def predict(self, X):
        predictions = np.column_stack([
            model.predict(X) for model in self.models_
        ])
        return np.mean(predictions, axis=1)   
```

According to the result of cross validation, we will choose some base models that behave very well as our input models.

```python
average_models=AveragingModels(models=(ridge, lasso, svr, gbdt))

score = cv_score(average_models)
print(" Averaged base models score: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
```

From models score, it is good te see that average_models improve our prediction scores. This will encourage us to look for more complex models to solve this problems.

* Complex stacking models

About complex stacking models, I have tried a lot of methods. What impressed me is a very convenient python package [mlxtend](#https://rasbt.github.io/mlxtend/). Therefore, I decide to choose it as my ways of stacking models. For more theories about stacking models, you can look up this package and it introduces very specifically.

For comparison, 'ridge','gbdt','svr','lasso' will be used as base models. Then we decide to choose respectively 'lgbr' and 'xgbr' as meta models. Let's have a look at the results below.

```python
stregr1 = StackingRegressor(regressors=[ridge, gbdt, svr,lasso], 
                           meta_regressor=xgbr)

score1 = cv_score(stregr1)
print("Stacking Averaged models score: {:.4f} ({:.4f})".format(score1.mean(), score1.std()))
```

```python
stregr2 = StackingRegressor(regressors=[ridge, gbdt, svr,lasso], 
                           meta_regressor=lgbr)

score2 = cv_score(stregr2)
print("Stacking Averaged models score: {:.4f} ({:.4f})".format(score2.mean(), score2.std()))
```
